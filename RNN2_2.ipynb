# Task Prediction

## Data Preparation

````python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
````

## Mapping tasks to indices and vice-versa

````python
tasks = ["chopping", "boiling", "strains", "smashing", "soup"]
task_to_idx = {task: idx for idx, task in enumerate(tasks)}
idx_to_task = {idx: task for idx, task in enumerate(tasks)}
````

## Input and Output Data

````python
X_tasks = [[task_to_idx[task]] for task in ["chopping", "boiling", "strains", "smashing", "soup"]]
y_tasks = [[task_to_idx[task]] for task in ["boiling", "strains", "smashing", "soup", "chopping"]]
````

## Converting to numpy arrays

````python
X_tasks = np.array(X_tasks)
y_tasks = np.array(y_tasks)
print(y_tasks)
````

## RNN Model Architecture

### Definition

````python
model_tasks = Sequential()
````

### Adding the RNN layer

````python
model_tasks.add(SimpleRNN(10, input_shape=(1, 1), return_sequences=True))
````

### Dense layer to map outputs to task space

````python
model_tasks.add(Dense(len(tasks), activation='softmax'))
````

## Model Compilation

````python
model_tasks.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
````

## Training the RNN Model



```
Try epochs=500 and epochs=1000 and comment the results
```



````python
model_tasks.fit(X_tasks, y_tasks, epochs=500)
````

## Generating task Sequence

````python
def generate_task_sequence(model, start_task, length):
    sequence = [start_task]
    input_task = start_task

    for _ in range(length):
        # Convertendo para representao numrica
        input_idx = task_to_idx[input_task]
        input_vector = np.array([[input_idx]])

        # Fazendo a previso
        prediction = model.predict(input_vector)
        next_idx = np.argmax(prediction)
        next_task = idx_to_task[next_idx]

        sequence.append(next_task)
        input_task = next_task

    return ' '.join(sequence)

print(generate_task_sequence(model_tasks, 'chopping', 10))
````
